Step 3: Handling Missing Values

```python
#Identifying missing values in columns

print(df.isnull().sum())

```

```python
#filliing missing values with 0
df_filled=df.fillna(0)

#Droping rows that have missing values
df_cleaned=df.dropna()

#Missing values after dropping
print(df_cleaned.isnull().sum())
```

Step 4:basic operations (filter rows, select columns)

```python
#Selecting the specific columns
selected_columns = df_cleaned[['title', 'final_price']]
print("Selected Columns")
display(selected_columns.head())
```

```python
#Filtering rows based on condition
filtered_df = df_cleaned[df_cleaned['rating'] > 4.0]
print(f"Number of rows after filtering: {len(filtered_df)}")
display(filtered_df.head())
```

Step 5: Remove Duplicates

```python
#to check how many duplicate rows exist in the dataset
duplicate_count = df_cleaned.duplicated().sum()
print(f"Duplicates found: {duplicate_count}")
```

```python
#Drop duplicate rows
df_no_duplicates = df_cleaned.drop_duplicates()
```

step 6:Create a derived column

```python
#adding a quantity column and giving every item a quantity of 2
df_no_duplicates['quantity'] = 2

# Create total_amount column by multiplying price and quantity
df_no_duplicates['total_amount'] = df_no_duplicates['initial_price'] * df_no_duplicates['quantity']

# 3. Print rows to see new columns

print(df_no_duplicates[['initial_price', 'quantity', 'total_amount']].head())
```

step 7: Saving the cleaned dataset to a new csv file

```python


# We use index=False so Pandas doesn't write the row numbers into the file
df_no_duplicates.to_csv('Cleaned_Shopping_dataset.csv', index=False)

print(" Cleaned dataset is saved successfully ")
```
